## Agregar datos por usuario

[Ahora que los datos están limpios, agrega los datos por usuario y por periodo para que solo haya un registro por usuario y por periodo. Esto facilitará mucho el análisis posterior.]

```python
# Calcula el número de llamadas hechas por cada usuario al mes. Guarda el resultado.
calls_per_user = calls.groupby(['user_id', 'month']).agg({
    'id': 'count',  # número de llamadas
    'duration': 'sum'  # minutos totales
}).reset_index()
calls_per_user.columns = ['user_id', 'month', 'total_calls', 'total_minutes']
print("Llamadas por usuario:")
print(calls_per_user.head())
```

```python
# Calcula la cantidad de minutos usados por cada usuario al mes. Guarda el resultado.
# Calculado en el paso anterior
```

```python
# Calcula el número de mensajes enviados por cada usuario al mes. Guarda el resultado.
messages_per_user = messages.groupby(['user_id', 'month']).agg({
    'id': 'count'
}).reset_index()
messages_per_user.columns = ['user_id', 'month', 'total_messages']
print("Mensajes por usuario:")
print(messages_per_user.head())
```

```python
# Calcula el volumen del tráfico de Internet usado por cada usuario al mes. Guarda el resultado.
internet_per_user = internet.groupby(['user_id', 'month']).agg({
    'mb_used': 'sum'
}).reset_index()
internet_per_user.columns = ['user_id', 'month', 'total_mb_used']
print("Internet por usuario:")
print(internet_per_user.head())
```

<div class="alert alert-block alert-success">
<b>Comentario de la revisora Iteración #1</b> <a class="tocSkip"></a>

Excelente uso del `groupby` para los diferentes dataframes y muy buen uso de `rename` para guardar los datos calculados en una nueva columna. </div>

[Junta los datos agregados en un DataFrame para que haya un registro que represente lo que consumió un usuario único en un mes determinado.]

```python
# Fusiona los datos de llamadas, minutos, mensajes e Internet con base en user_id y month
user_behavior = users[['user_id', 'city', 'plan']].copy()

# Crear DataFrame para todos los meses
all_months_data = pd.DataFrame()
for month in range(1, 13):
    month_data = user_behavior.copy()
    month_data['month'] = month
    
    # Fusionar con llamadas
    month_calls = calls_per_user[calls_per_user['month'] == month]
    month_data = pd.merge(month_data, month_calls, on=['user_id', 'month'], how='left')
    
    # Fusionar con mensajes
    month_messages = messages_per_user[messages_per_user['month'] == month]
    month_data = pd.merge(month_data, month_messages, on=['user_id', 'month'], how='left')
    
    # Fusionar con internet
    month_internet = internet_per_user[internet_per_user['month'] == month]
    month_data = pd.merge(month_data, month_internet, on=['user_id', 'month'], how='left')
    
    all_months_data = pd.concat([all_months_data, month_data])

# Llenar valores nulos con 0
all_months_data = all_months_data.fillna({
    'total_calls': 0,
    'total_minutes': 0,
    'total_messages': 0,
    'total_mb_used': 0
})

print("Datos fusionados:")
print(all_months_data.tail())
```

```python
# Añade la información de la tarifa
plan_details = plans.set_index('plan_name')
all_months_data = all_months_data.merge(plan_details, left_on='plan', right_index=True, how='left')
print("Datos con información de tarifa:")
print(all_months_data.head())
```

[Calcula los ingresos mensuales por usuario (resta el límite del paquete gratuito del número total de llamadas, mensajes de texto y datos; multiplica el resultado por el valor del plan de llamadas; añade la tarifa mensual en función del plan de llamadas). Nota: Dadas las condiciones del plan, ¡esto podría no ser tan trivial como un par de líneas! Así que no pasa nada si dedicas algo de tiempo a ello.]

```python
# Calcula el ingreso mensual para cada usuario
# Calcula el ingreso mensual para cada usuario
def calculate_monthly_revenue(row):
    monthly_fee = row['usd_monthly_pay']
    
    # Llamadas
    minutes_used = row['total_minutes']
    minutes_included = row['minutes_included']
    cost_per_minute = row['usd_per_minute']
    
    if minutes_used > minutes_included:
        call_cost = (minutes_used - minutes_included) * cost_per_minute
    else:
        call_cost = 0
    
    # Mensajes
    messages_used = row['total_messages']
    messages_included = row['messages_included']
    cost_per_message = row['usd_per_message']
    
    if messages_used > messages_included:
        message_cost = (messages_used - messages_included) * cost_per_message
    else:
        message_cost = 0
    
    # Datos (convertir MB a GB y redondear hacia arriba)
    mb_used = row['total_mb_used']
    mb_included = row['mb_per_month_included']
    cost_per_gb = row['usd_per_gb']
    
    if mb_used > mb_included:
        extra_gb = np.ceil((mb_used - mb_included) / 1024)
        data_cost = extra_gb * cost_per_gb
    else:
        data_cost = 0
    
    total_revenue = monthly_fee + call_cost + message_cost + data_cost
    return total_revenue

# Calcular ingresos mensuales
all_months_data['monthly_revenue'] = all_months_data.apply(calculate_monthly_revenue, axis=1)
print("Datos con ingresos calculados:")
print(all_months_data[['user_id', 'month', 'plan', 'total_minutes', 'total_messages', 'total_mb_used', 'monthly_revenue']].head())
```

Hiciste un muy buen trabajo al manejar los sobrecostos con una instancia condicional para evitar valores negativos par alos usuarios que no tuvieron sobrecostos. Buen trabajo de redondeo de las GB de internet con `np.ceil`.

## Prueba las hipótesis estadísticas

[Prueba la hipótesis de que son diferentes los ingresos promedio procedentes de los usuarios de los planes de llamada Ultimate y Surf.]

[Elabora las hipótesis nula y alternativa, escoge la prueba estadística, determina el valor alfa.]

```python
# Prueba las hipótesis
# Preparar datos para la prueba
surf_revenue = all_months_data[all_months_data['plan'] == 'surf']['monthly_revenue']
ultimate_revenue = all_months_data[all_months_data['plan'] == 'ultimate']['monthly_revenue']

print(f"Tamaño muestra Surf: {len(surf_revenue)}")
print(f"Tamaño muestra Ultimate: {len(ultimate_revenue)}")
print(f"Media Surf: ${surf_revenue.mean():.2f}")
print(f"Media Ultimate: ${ultimate_revenue.mean():.2f}")
t_stat, p_value = stats.ttest_ind(surf_revenue, ultimate_revenue, equal_var=False)
print(f"\nPrueba t (varianzas desiguales):")
print(f"p-value = {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print(f"\nCon un nivel de significancia de {alpha}, rechazamos H0.")
    print("Existe evidencia estadística de que los ingresos promedio difieren entre los planes Surf y Ultimate.")
else:
    print(f"\nCon un nivel de significancia de {alpha}, no rechazamos H0.")
    print("No existe evidencia estadística de que los ingresos promedio difieran entre los planes.")
```

[Prueba la hipótesis de que el ingreso promedio de los usuarios del área NY-NJ es diferente al de los usuarios de otras regiones.]

[Elabora las hipótesis nula y alternativa, escoge la prueba estadística, determina el valor alfa.]

```python
# Prueba las hipótesis
# Identificar usuarios de NY-NJ
ny_nj_users = all_months_data[
    all_months_data['city'].str.contains('NY|NJ', na=False)
]
other_users = all_months_data[
    ~all_months_data['city'].str.contains('NY|NJ', na=False)
]

print(f"Usuarios NY-NJ: {ny_nj_users['user_id'].nunique()}")
print(f"Usuarios otras regiones: {other_users['user_id'].nunique()}")

ny_nj_revenue = ny_nj_users['monthly_revenue']
other_revenue = other_users['monthly_revenue']

print(f"\nMedia NY-NJ: ${ny_nj_revenue.mean():.2f}")
print(f"Media otras regiones: ${other_revenue.mean():.2f}")

# Prueba t para muestras independientes
t_stat2, p_value2 = stats.ttest_ind(ny_nj_revenue, other_revenue, equal_var=False)
print(f"\nPrueba t (varianzas desiguales):")
print(f"t-statistic = {t_stat2:.4f}, p-value = {p_value2:.4f}")

# Hipótesis:
# H0: μ_ny_nj = μ_other (no hay diferencia en ingresos promedio)
# H1: μ_ny_nj ≠ μ_other (hay diferencia en ingresos promedio)

if p_value2 < alpha:
    print(f"\nCon un nivel de significancia de {alpha}, rechazamos H0.")
    print("Existe evidencia estadística de que los ingresos promedio difieren entre NY-NJ y otras regiones.")
else:
    print(f"\nCon un nivel de significancia de {alpha}, no rechazamos H0.")
    print("No existe evidencia estadística de que los ingresos promedio difieran entre NY-NJ y otras regiones.")
```
